```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes_split.csv')
```

```python
# df
```

```python
# df = pd.read_csv('diabetes_012_health_indicators_BRFSS2015.csv')
```

```python
df['Diabetes_binary'] = df['Diabetes_binary'].replace(2, 1)
```

```python
df['BMI'] = df['BMI'] / 98.0
df['GenHlth'] = df['GenHlth'] / 5.0
df['MentHlth'] = df['MentHlth'] / 30.0
df['PhysHlth'] = df['PhysHlth'] / 30.0
df['Age'] = df['Age'] / 13.0
df['Education'] = df['Education'] / 6.0
df['Income'] = df['Income'] / 8.0
```

```python
X = df.drop('Diabetes_binary', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=4)
```

```python
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
```

```python
accuracy_score(lda.predict(X_train), y_train)
```

```python
accuracy_score((lda).predict(X_test),y_test)
```

```python
from sklearn.metrics import confusion_matrix

print("Confusion Matrix")
print(confusion_matrix(y_test, lda.predict(X_test)))
```

```python
randomForest = RandomForestClassifier(n_estimators=200, max_depth=50, min_samples_leaf=10, min_samples_split=2, random_state=4, oob_score=True)
randomForest.fit(X_train, y_train)
```

```python
randomForest2 = RandomForestClassifier(n_estimators=100, max_depth=50, random_state=4, oob_score=True)
randomForest2.fit(X_train, y_train)
```

```python
randomForest3 = RandomForestClassifier(n_estimators=20, max_depth=15, random_state=4, oob_score=True)
randomForest3.fit(X_train, y_train)
```

```python
accuracy_score(randomForest.predict(X_train).round(), y_train)
```

```python
accuracy_score(randomForest.predict(X_test).round(),y_test)
```

```python
decisionTree = DecisionTreeClassifier(max_depth=100, max_features='sqrt', min_samples_leaf=10, random_state=4)
decisionTree.fit(X_train, y_train)
```

```python
accuracy_score(decisionTree.predict(X_train), y_train)
```

```python
accuracy_score(decisionTree.predict(X_test),y_test)
```

```python
from sklearn.metrics import confusion_matrix

print("Confusion Matrix")
print(confusion_matrix(y_test, decisionTree.predict(X_test)))
```

```python
import numpy as np
import joblib

feature_label = np.array(['HighBP', 'HighChol', 'CholCheck', 'BMI', 'Smoker', 'Stroke', 'HeartDiseaseorAttack', 'PhysActivity', 'Fruits', 'Veggies', 'HvyAlcoholConsump', 'AnyHealthCare', 'NoDocbcCost', 'GenHlth', 'MentHlth', 'PhysHlth', 'DiffWalk', 'Sex', 'Age', 'Education', 'Income', ])

model = joblib.load('randomforest.pkl')

importances = model.feature_importances_

feature_indexes_by_importance = importances.argsort()

for index in feature_indexes_by_importance:
    print("{} - {:.2f}%".format(feature_label[index], (importances[index] * 100.0)))
```
